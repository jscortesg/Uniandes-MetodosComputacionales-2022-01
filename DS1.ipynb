#Entrenamiento y separación de datos

Librería: **scikit learn**

```python
from sklearn.datasets import load_iris

iris_dataset = load_iris()
```

```python
type(iris_dataset)
```

```python
iris_dataset
```

```python
iris_dataset.keys()
```

```python
iris_dataset['target_names']
```

```python
iris_dataset['feature_names']
```

```python
iris_dataset['data'][:5]
```

```python
iris_dataset['target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state = 0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Tabla de datos

```python
import pandas as pd
```

```python
iris_dataframe = pd.DataFrame(X_train, columns = iris_dataset.feature_names)
```

```python
iris_dataframe.head(10)
```

```python
iris_dataframe.corr()
```

```python
!pip install mglearn
import mglearn

grr = pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='o', 
                        hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

Predicción

```python
import numpy as np

X_new = np.array([[5, 2.9, 1, 0.2]])
```

```python
prediction = knn.predict(X_new)
prediction
```

```python
print("Objetivo predicho: {}".format(iris_dataset['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print("Predicción para X_test:\n {}".format(y_pred))
```

```python
print("Precisión de la predicción: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
print("Score de X_test: {:.2f}".format(knn.score(X_test, y_test)))
```

```python
knn2 = KNeighborsClassifier(n_neighbors=2)
```

```python
knn2.fit(X_train, y_train)
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
prediction2 = knn2.predict(X_new)

print("Objetivo predicho: {}".format(iris_dataset['target_names'][prediction2]))
```

```python
y_pred2 = knn2.predict(X_test)
print("Predicción para X_test:\n {}".format(y_pred2))
```

```python
print("Score de X_test: {:.2f}".format(knn2.score(X_test, y_test)))
```

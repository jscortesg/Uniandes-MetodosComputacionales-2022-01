#Análisis de Fourier

Series de Fourier:

Para toda función periódica $y(t)$ tal que $y(t + T) = y(t)$, con una frecuencia angular fundamental

\begin{equation}
\omega = \omega_{1} = \frac{2\pi}{T},
\end{equation}

y $T$ el periodo, la serie de Fourier se escribe como 

\begin{equation}
y(t) = a_{0} + \sum_{n=1}^{\infty}\left[a_{n}\cos(n\omega_{1}t) + b_{n}\sin(n\omega_{1}t)\right].
\end{equation}

Base de funciones ortogonales en $(0\leq t\leq T)$: $(1,\cos(n\omega_{1}t),\sin(n\omega_{1}t))$.

\begin{align}
&\int_{0}^{T}1\cdot\sin(n\omega_{1} t)\,dt = \int_{0}^{T}1\cdot\cos(n\omega_{1} t)\,dt = 0  \\
&\int_{0}^{T}\sin(n\omega_{1} t)\cos(n\omega_{1} t)\,dt = 0\\
&\int_{0}^{T}\sin(n\omega_{1} t)\sin(m\omega_{1}t)\,dt = \frac{T}{2}\delta_{nm} = 
\begin{cases}
T/2, & n=m \\
0, & n\neq m
\end{cases} \\
&\int_{0}^{T}\cos(n\omega_{1} t)\cos(m\omega_{1}t)\,dt =\frac{T}{2}\delta_{nm} =
\begin{cases}
T/2, & n=m \\
0, & n\neq m
\end{cases}
\end{align}

Coeficientes de la expansión:

Por propiedades de ortogonalidad,

\begin{align}
\int_{0}^{T}1\cdot y(t)\,dt &= \int_{0}^{T}a_{0}\,dt + \sum_{n=1}^{\infty}\left[a_{n}\int_{0}^{T}1\cdot\cos(n\omega_{1} t)\,dt + b_{n}\int_{0}^{T}1\cdot\sin(n\omega_{1} t)\,dt\right] \\
a_{0} &= \frac{1}{T}\int_{0}^{T}y(t)\,dt \\
& \\
\int_{0}^{T}\cos(m\omega_{1} t)\cdot y(t)\,dt &= a_{0}\int_{0}^{T}1\cdot\cos(m\omega_{1} t)\,dt + \sum_{n=1}^{\infty}\left[a_{n}\int_{0}^{T}\cos(n\omega_{1} t)\cdot\cos(m\omega_{1} t)\,dt + b_{n}\int_{0}^{T}\cos(m\omega_{1} t)\cdot\sin(n\omega_{1} t)\,dt\right] \\
\int_{0}^{t}y(t)\cos(m\omega_{1} t)\,dt &= \sum_{n=1}^{\infty}a_{n}*\frac{T}{2}\delta_{nm} = a_{m}\frac{T}{2} \\
a_{m} &= \frac{2}{T}\int_{0}^{T}y(t)\cos(m\omega_{1} t)\,dt \\
& \\
\int_{0}^{T}\sin(m\omega_{1} t)\cdot y(t)\,dt &= a_{0}\int_{0}^{T}1\cdot\sin(m\omega_{1} t)\,dt + \sum_{n=1}^{\infty}\left[a_{n}\int_{0}^{T}\cos(n\omega_{1} t)\cdot\sin(m\omega_{1} t)\,dt + b_{n}\int_{0}^{T}\sin(n\omega_{1} t)\cdot\sin(m\omega_{1} t)\,dt\right] \\
\int_{0}^{T}y(t)\sin(m\omega_{1} t)\,dt &= \sum_{n=1}^{\infty}b_{n}*\frac{T}{2}\delta_{nm} = b_{m}\frac{T}{2} \\
b_{m} &= \frac{2}{T}\int_{0}^{T}y(t)\sin(m\omega_{1} t)\,dt
\end{align}


Si $y(t)$ es par ($y(t) = y(-t)$), 

\begin{equation}
b_{m}=0.
\end{equation}

Si $y(t)$ es impar ($y(t) = -y(-t)$), 

\begin{equation}
a_{m}=0.
\end{equation}

Ejemplo: señal cuadrada

\begin{equation}
y(t) = 
\begin{cases}
1, & 0\leq t\leq T/2 \\
-1, & T/2\leq t \leq T
\end{cases}
\end{equation}

Coeficientes:

\begin{align}
a_{0}&=\frac{1}{T}\int_{0}^{T}y(t)\,dt=\frac{1}{T}\int_{0}^{T/2}dt-\frac{1}{T}\int_{T/2}^{T}dt = \frac{1}{2}-\frac{1}{2} = 0 \\
a_{n}&=\frac{2}{T}\int_{0}^{T}y(t)\cos(n\omega_{1}t)\,dt = \frac{2}{T}\int_{0}^{T}y(t)\cos\left(\frac{2\pi n}{T}t\right)\,dt = \frac{2}{T}\left[\int_{0}^{T/2}\cos\left(\frac{2\pi n}{T}t\right)\,dt - \int_{T/2}^{T}\cos\left(\frac{2\pi n}{T}t\right)\,dt\right] \\
&= \frac{2}{T}*\frac{T}{2\pi n}\left[\left. \sin\left(\frac{2\pi n}{T}t\right)\right|_{0}^{T/2} - \left. \sin\left(\frac{2\pi n}{T}t\right)\right|_{T/2}^{T}\right] = 0 \\
b_{n}&=\frac{2}{T}\int_{0}^{T}y(t)\sin(n\omega_{1}t)\,dt = \frac{2}{T}\int_{0}^{T}y(t)\sin\left(\frac{2\pi n}{T}t\right)\,dt = \frac{2}{T}\left[\int_{0}^{T/2}\sin\left(\frac{2\pi n}{T}t\right)\,dt - \int_{T/2}^{T}\sin\left(\frac{2\pi n}{T}t\right)\,dt\right] \\
&= -\frac{2}{T}*\frac{T}{2\pi n}\left[\left. \cos\left(\frac{2\pi n}{T}t\right)\right|_{0}^{T/2} - \left. \cos\left(\frac{2\pi n}{T}t\right)\right|_{T/2}^{T}\right] = -\frac{1}{n\pi}[\cos(n\pi)-1 - 1 + \cos(n\pi)] \\
b_{n} &= \frac{2}{n\pi}\left[1-(-1)^{n}\right]
\end{align}

Entonces,

\begin{equation}
y(t) = \frac{2}{\pi}\sum_{n=1}^{\infty}\frac{\left[1-(-1)^{n}\right]}{n}\sin\left(\frac{2\pi n}{T}t\right).
\end{equation}

```python
def serie_oc(t, n, Pe):

  def coef_oc(n):
    return (1-pow(-1,n))/n  

  def sen_oc(t, n, Pe):
    return np.sin((2*np.pi*n*t)/Pe)
    
  return (2/np.pi)*coef_oc(n)*sen_oc(t, n, Pe)
```

Generalización a la representación compleja:

$\cos x = \frac{e^{i x}+e^{-i x}}{2}$. $\sin x = \frac{e^{i x}-e^{-i x}}{2i}$.

\begin{align}
y(t) & = a_{0} + \sum_{n=1}^{\infty}\left[\frac{a_{n}}{2}\left(e^{2\pi i n*t/T} + e^{-2\pi i n*t/T}\right) + \frac{b_{n}}{2i}\left(e^{2\pi i n*t/T} - e^{-2\pi i n*t/T}\right)\right] \\
& = a_{0}+\sum_{n=1}^{\infty}\left(\frac{a_{n}-ib_{n}}{2}\right)e^{2\pi in*t/T} + \sum_{n=1}^{\infty}\left(\frac{a_{n}+ib_{n}}{2}\right)e^{-2\pi in*t/T}
\end{align}

Coeficientes: 

\begin{equation}
c_{n} = \frac{a_{n}-ib_{n}}{2}.
\end{equation}

Por definición, $c_{0} = \frac{a_{0}-ib_{0}}{2}=a_{0}$. Además, 

\begin{equation}
\sum_{n=1}^{\infty}\left(\frac{a_{n}+ib_{n}}{2}\right)e^{-2\pi in*t/T} = \sum_{n=-1}^{\infty}\left(\frac{a_{-n}+ib_{-n}}{2}\right)e^{2\pi in*t/T}.
\end{equation}

Finalmente,

\begin{equation}
y(t) = \sum_{n=-\infty}^{\infty}c_{n}\exp\left(\frac{2\pi i n}{T}t\right)
\end{equation}

Coeficientes:

\begin{align}
c_{n} &= \frac{1}{2}(a_{n}-ib_{n}) = \frac{1}{2}*\frac{2}{T}\int_{0}^{T}y(t)\left[\cos\left(\frac{2\pi n}{T}t\right)-i\sin\left(\frac{2\pi n}{T}t\right)\right]dt \\
\cos x -i\sin x &= \frac{e^{ix}+e^{-ix}}{2} - \frac{e^{ix}-e^{-ix}}{2} = e^{-ix} \\
c_{n} &= \frac{1}{T}\int_{0}^{T}y(t)e^{-2\pi in*t/T}\,dt \\
\text{O también, }& \\
c_{n} &= \frac{1}{T}\int_{-T/2}^{T/2}y(t)e^{-2\pi in*t/T}\,dt
\end{align}

#Transformada de Fourier

Para funciones periódicas,

\begin{equation}
y(t) = \sum_{n=-\infty}^{\infty}c_{n}\exp\left(\frac{2\pi i n}{T}t\right)
\end{equation}

Para funciones NO periódicas, es decir, $y(t)\neq y(t+T)$, 

\begin{equation}
\sum_{n=-\infty}^{\infty}c_{n}\rightarrow \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty}d\omega Y(\omega),
\end{equation}

$\omega$: continuo de frecuencias.

Entonces,

\begin{equation}
y(t)=\frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty}Y(\omega)e^{i\omega t}\,d\omega,
\end{equation}

donde $Y(\omega)$: transformada de Fourier de $y(t)$.

En términos de una integral,

\begin{equation}
Y(\omega) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty}y(t)e^{-i\omega t}\,dt
\end{equation}

#Transformada Discreta Finita de Fourier

\begin{equation}
Y(\omega) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty}y(t)e^{-i\omega t}\,dt
\end{equation}

Señal medida en intervalos de tiempo $\Delta t$ uniformes:

\begin{align}
&\Delta t = h, \\
&t_{k} = kh,\, k=1,2,3,\cdots,N \\
&t_{1} = h, \,t_{N}=Nh \\
&y(t_{k})\equiv y_{k}.
\end{align}

Señal: conocida en $0\leq t\leq T$.

\begin{equation}
Y(\omega_{n})=\frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty}y(t)e^{-i\omega_{n} t}\,dt\approx \frac{1}{\sqrt{2\pi}}\int_{0}^{T}y(t)e^{-i\omega_{n} t}\,dt 
\end{equation}

Usando regla del trapecio,

\begin{align}
\int_{0}^{T}y(t)e^{-i\omega_{n} t}\,dt&\approx\frac{h}{2}\sum_{k=0}^{N-1}\left[y(t_{k})+y(t_{k+1})\right]e^{-i\omega_{n} t} \\
\sum_{k=0}^{N-1}\left[y(t_{k})+y(t_{k+1})\right]&=y(t_{0})+y(t_{1}) + y(t_{1}) \\
&+ y(t_{2}) + y(t_{2}) + y(t_{3})+y(t_{3})+\cdots \\
& +y(t_{N-1}) + y(t_{N-1})+ y(t_{N}) \\
&=y(t_{0})+2\sum_{k=1}^{N-1}y(t_{k})+y(t_{N}) = y_{0} + 2\sum_{k=1}^{N-1}y_{k} + y_{N}
\end{align}

Añadiendo periodicidad, es decir, $y_{0} = y_{N}$, $y_{0} + 2\sum_{k=1}^{N-1}y_{k} + y_{N} = 2y_{N} + 2\sum_{k=1}^{N-1}y_{k} = 2\sum_{k=1}^{N}y_{k}$. 

Entonces, en un periodo $T = Nh$,

\begin{align}
Y(\omega_{n}) &= h\sum_{k=1}^{N}y_{k}\frac{e^{-i\omega_{n}t_{k}}}{\sqrt{2\pi}}, \\
t_{k} &= kh, \, \omega_{n} = n\omega_{1} = n\frac{2\pi}{T} = n\frac{2\pi}{Nh}, \\
Y(\omega_{n})&=h\sum_{k=1}^{N}y_{k}\frac{e^{-2\pi i k n/N}}{\sqrt{2\pi}} \\
&\text{O también,} \\
Y_{n}&\equiv \frac{Y(\omega_{n})}{h} = \sum_{k=1}^{N}y_{k}\frac{e^{-2\pi i kn/N}}{\sqrt{2\pi}.}
\end{align}


Transformada inversa:

\begin{align}
y(t) &= \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty}Y(\omega)e^{i\omega t}\,d\omega \\
d\omega & \approx \Delta \omega = \frac{2\pi}{N h}\\
y(t) &\approx \sum_{n=1}^{N}\frac{2\pi}{\sqrt{2\pi}Nh}Y(\omega_{n})e^{i\omega_{n}t} \\
y(t)&\approx\sum_{n=1}^{N}\frac{2\pi}{N}\frac{e^{i\omega_{n}t}}{\sqrt{2\pi}}Y_{n}.
\end{align}

Cantidad de operaciones: $N$ sumas y $N$ multiplicaciones = $N*N=\mathcal{O}(N^{2})$.

```python
import matplotlib.pyplot as plt
import numpy as np
import math

plt.style.use('seaborn-poster')
%matplotlib inline
```

```python
# sampling rate
sr = 100
# sampling interval
ts = 1.0/sr
t = np.arange(0,1,ts)

freq = 1.
x = 3*np.sin(2*np.pi*freq*t)

freq = 4
x += np.sin(2*np.pi*freq*t)

freq = 7   
x += 0.5* np.sin(2*np.pi*freq*t)

plt.figure(figsize = (8, 6))
plt.plot(t, x, 'r')
plt.ylabel('Amplitude')

plt.show()
```

```python
def DFT(x):
    """
    Function to calculate the 
    discrete Fourier Transform 
    of a 1D real-valued signal x
    """

    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    
    X = np.dot(e, x)/math.sqrt(2*np.pi)
    
    return X
```

```python
DFT(x)
```

```python
X = DFT(x)

# calculate the frequency
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 

plt.figure(figsize = (8, 6))
plt.stem(freq, abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')
plt.show()
```

Transformada Rápida de Fourier (Fast Fourier Transform -FFT-):

La DFT es simétrica, es decir,

\begin{align}
Y_{n+N} &= \sum_{k=1}^{N}y_{k}\frac{e^{-2\pi i k (n+N)/N}}{\sqrt{2\pi}} = \sum_{k=1}^{N}y_{k}\frac{e^{-2\pi i k n/N}}{\sqrt{2\pi}}e^{-2\pi i k} \\
e^{-2\pi i k} &= 1, \,k=1, 2, 3,\cdots \\
Y_{n+N} &= \sum_{k=1}^{N}y_{k}\frac{e^{-2\pi i k n/N}}{\sqrt{2\pi}} = Y_{n}.
\end{align}

Por otro lado,

\begin{align}
Y_{n}&\equiv \frac{Y(\omega_{n})}{h} = \sum_{k=1}^{N}y_{k}\frac{e^{-2\pi i kn/N}}{\sqrt{2\pi}.} \\
Y_{n} &= Y_{2n} + Y_{2n+1} \\
Y_{n} &= \sum_{k=1}^{N/2}y_{2k}\frac{e^{(-2\pi i *2*k*n)/N}}{\sqrt{2\pi}} + \sum_{k=1}^{N/2}y_{2k+1}\frac{e^{[-2\pi i n*(2k+1)]/N}}{\sqrt{2\pi}} \\
\frac{e^{[-2\pi i n*(2k+1)]/N}}{\sqrt{2\pi}}& = \frac{e^{(-2\pi i k*2n)/N}}{\sqrt{2\pi}}*e^{-2\pi i n/N} = e^{-2\pi i n/N}*\frac{e^{-2\pi i kn/(N/2)}}{\sqrt{2\pi}} \\
Y_{n}& = \sum_{k=1}^{N/2}y_{2k}\frac{e^{-2\pi i kn/(N/2)}}{\sqrt{2\pi}} + e^{-2\pi i n/N}\sum_{k=1}^{N/2}y_{2k+1}\frac{e^{-2\pi i kn/(N/2)}}{\sqrt{2\pi}}
\end{align}

Cantidad de operaciones: $\mathcal{O}(N\log_{2} N)$.

```python
from numpy.fft import fft, ifft

X = fft(x)/math.sqrt(2*np.pi)
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 

plt.figure(figsize = (15, 6))
plt.subplot(131)

plt.stem(freq, np.abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude |X(freq)|')
plt.xlim(0, 100)

plt.subplot(132)

plt.stem(freq, np.abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude |X(freq)|')
plt.xlim(0, 10)


plt.subplot(133)
plt.plot(t, math.sqrt(2*np.pi)*ifft(X), 'r')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```

```python
Y2 = fft(y2)/math.sqrt(2*np.pi)
N2 = len(Y2)
n2 = np.arange(N2)
T2 = N2/sr
freq2 = n2/T2 


plt.figure(figsize = (12, 6))
plt.subplot(121)

plt.stem(freq2, np.abs(Y2), 'r', \
         markerfmt=" ", basefmt="-r")
plt.xlabel('f (Hz)')
plt.ylabel('Amplitud FFT de $y_{2}(t)$')
plt.xlim(0, 200)

plt.subplot(122)

plt.stem(freq2, np.abs(Y2), 'r', \
         markerfmt=" ", basefmt="-r")
plt.xlabel('f (Hz)')
plt.title('Amplitud FFT de $y_{2}(t)$')
plt.xlim(0, 20)
```

#Ecuaciones Diferenciales Ordinarias (EDO)

Forma general (orden $n$):

\begin{equation}
F\left(x,\,f(x),\,\frac{df}{dx},\,\frac{d^{2}f}{dx^{2}},\cdots,\frac{d^{n-1}f}{dx^{n-1}}\right) = \frac{d^{n}f}{dx^{n}}.
\end{equation}

Condiciones iniciales (valores conocidos): 

\begin{equation}
f(x_{0})\,,\left. \frac{df}{dx}\right|_{x=x_{0}}\,,\left. \frac{d^{2}f}{dx^{2}}\right|_{x=x_{0}}\,,\cdots,\left. \frac{d^{n-1}f}{dx^{n-1}}\right|_{x=x_{0}}
\end{equation}

Ejemplo: movimiento armónico simple amortiguado de la forma

\begin{align}
\text{EDO asociada: }&m\frac{d^{2}x}{dt^{2}} = -kx - b\frac{dx}{dt}. \\
\text{Condiciones iniciales: }&x(0) = x_{0}, \, v(0) = 0. \\
\text{Solución: }&x(t) = Ae^{-\frac{\gamma}{2}t}\cos(\omega t + \phi)
\end{align}

Ejemplo de **EDO lineal y homogénea**.

Potencias de $\frac{d^{j-1}f}{dx^{j-1}},\,j=1,\cdots,n$: EDO **no lineal**.

Como por ejemplo,

\begin{equation}
\frac{dy}{dt} = -3t^{2}y + t^{9} +y^{7}
\end{equation}

EDO de primer orden:

\begin{equation}
\frac{dy}{dt} = f(t,y).
\end{equation}

Descarga de un condensador en un circuito RC:

\begin{align}
&RI+\frac{Q}{C} = 0,\,I = \frac{dQ}{dt} \\
&\frac{dQ}{dt} + \frac{1}{RC}Q = 0. \\
&\text{Condición inicial: }Q(0) = Q_{0}.
\end{align}


#Reducción del orden en la EDO

Recordando que 

\begin{equation}
F\left(x,\,f(x),\,\frac{df}{dx},\,\frac{d^{2}f}{dx^{2}},\cdots,\frac{d^{n-1}f}{dx^{n-1}}\right) = \frac{d^{n}f}{dx^{n}},
\end{equation}

o de igual forma,

\begin{equation}
F\left(x,\,f(x),\,f^{(1)}(x),\,f^{(2)}(x),\cdots,f^{(n-1)}(x)\right) = f^{(n)}(x),
\end{equation}

sea el vector $S(x)$

\begin{equation}
S(x) =
\begin{pmatrix}
f(x) \\
f^{(1)}(x) \\
f^{(2)}(x) \\
\vdots \\
f^{(n-1)}(x)
\end{pmatrix}=
\begin{pmatrix}
S_{1}(x) \\
S_{2}(x) \\
S_{3}(x) \\
\vdots \\
S_{n}(x),
\end{pmatrix}
\end{equation}

donde $S_{n}(x) = f^{(n-1)}(x)$. De este modo,

\begin{equation}
\frac{dS}{dx} =
\begin{pmatrix}
f^{(1)}(x) \\
f^{(2)}(x) \\
f^{(3)}(x) \\
\vdots \\
f^{(n)}(x)
\end{pmatrix}=
\begin{pmatrix}
S_{2}(x) \\
S_{3}(x) \\
S_{4}(x) \\
\vdots \\
F(x,S_{1}(x),S_{2}(x),\cdots,S_{n-1}(x)).
\end{pmatrix}
\end{equation}

De manera más compacta,

\begin{equation}
\frac{dS}{dx} = \mathcal{F}(x,S(x)).
\end{equation}

#Método de Euler para EDO de 1er Order

Para una EDO tipo $\frac{dy}{dx} = F(x,y(x))$, sea $x={x_{1},x_{2},\cdots, x_{N}}$ con $x_{j+1}-x_{j}=h$. 

\begin{align}
\frac{dy}{dx}&\approx \left. \frac{dy}{dx}\right|_{\text{forward}} = \frac{y(x_{j+1})-y(x_{j})}{h} \\
\frac{y(x_{j+1})-y(x_{j})}{h} &= F(x_{j},y(x_{j})) \\
y(x_{j+1}) &= y(x_{j}) + hF(x_{j},y(x_{j})) \\
\text{Tomando }y(x_{j}) &= y_{j} \\
y_{j+1} &= y_{j} + hF(x_{j},y_{j})
\end{align}

**Método de Euler explícito**


El **método implícito** es dado por

\begin{align}
\frac{dy}{dx} &= F(x,y(x)) \\
\int_{y_{j}}^{y_{j+1}}dy &= \int_{x_{j}}^{x_{j+1}}F(x,y(x))\,dx \\
\int_{x_{j}}^{x_{j+1}}F(x,y(x))\,dx &\approx (x_{j+1}-x_{j})F(x_{j+1},y(x_{j+1})) = hF(x_{j+1},y_{j+1}),\,h=x_{j+1}-x_{j} \\
y_{j+1} - y_{j} &= hF(x_{j+1},y_{j+1}) \\
y_{j+1} &= y_{j} + hF(x_{j+1},y_{j+1})
\end{align}

**Método trapezoidal:** $\frac{\text{método explícito} + \text{método implícito}}{2}$

\begin{equation}
y_{j+1} = y_{j} + \frac{h}{2}\left[F(x_{j},y(x_{j})) + F(x_{j+1},y(x_{j+1}))\right]
\end{equation}

**Método trapezoidal para ecuaciones con orden reducido**

De forma matricial,

\begin{align}
\frac{dS}{dx} = \mathcal{F}(x,y(x)).
\end{align}

Sea $\mathcal{F}(x,y(x)) = \mathcal{M}S$.

Método explícito:

\begin{align}
\frac{S_{j+1} - S_{j}}{h} &= \mathcal{M}S_{j} \\
S_{j+1} &= \left(1+h\mathcal{M}\right)S_{j}
\end{align}

Método implícito:

\begin{align}
\frac{S_{j+1} - S_{j}}{h} &= \mathcal{M}S_{j+1} \\
\left(1-h\mathcal{M}\right)S_{j+1} &= S_{j}
\end{align}

Método trapezoidal:

\begin{align}
\frac{1}{2}\left[S_{j+1} + \left(1-h\mathcal{M}\right)S_{j+1}\right] &= \frac{1}{2}\left[S_{j} + \left(1+h\mathcal{M}\right)S_{j}\right] \\
\left(1-\frac{h\mathcal{M}}{2}\right)S_{j+1} &= \left(1+\frac{h\mathcal{M}}{2}\right)S_{j} \\
S_{j+1} &= \left(1-\frac{h\mathcal{M}}{2}\right)^{-1}\left(1+\frac{h\mathcal{M}}{2}\right)S_{j}
\end{align}

Ejemplo 1: ecuación diferencial de la forma

\begin{equation}
\frac{dy}{dt} = e^{-t},\,y(0)=y_{0} = -1.
\end{equation}

Evaluar en intervalo $0\leq t \leq 1$.

Solución exacta:

\begin{align}
\frac{dy}{dt} &=e^{-t} \\
\int_{y_{0}}^{y}dy &= \int_{0}^{t}e^{-t}dt \\
y - y_{0} &= y - (-1) = -\left. e^{-t}\right|_{0}^{t} = 1 - e^{-t} \\
y(t) &= -e^{-t}.
\end{align}

Para este caso, 

\begin{align}
\frac{dy}{dt} &= F(t,y(t)) \\
F(t,y(t)) &= e^{-t}, \, y(0) = y_{0} = -1.
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# Define parameters
F = lambda t, y: np.exp(-t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
y0 = -1 # Initial Condition

# Explicit Euler Method
y = np.zeros(len(t))
y[0] = y0

for i in range(0, len(t) - 1):
    y[i + 1] = y[i] + h*F(t[i], y[i])

plt.figure(figsize = (12, 8))
plt.plot(t, y, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

Ejemplo 2: caída libre en $0\leq t\leq 1$ s.

\begin{align}
&\frac{d^{2}y}{dt^{2}} = -g \\
&v(0) = 0,\, y(0) = 10\text{ m}
\end{align}

Velocidad: $a = \frac{dv}{dt}$. Entonces, $\frac{dv}{dt} = -g$.

\begin{align}
\frac{dv}{dt} &\approx \frac{v_{j+1}-v_{j}}{h} = F(v_{j},t_{j}), \,F(v_{j},t_{j}) = -g \\
v_{j+1} &= v_{j} + hF(v_{j},t_{j})
\end{align}

Posición: $v = -gt = \frac{dy}{dt}$. De manera análoga, 

\begin{align}
-gt &= \frac{dy}{dt} \approx \frac{y_{j+1} - y_{j}}{h}. \\
\text{Sea }F(t,y(t)) &= -gt. \\
y_{j+1} &= y_{j} + F(t_{j},y_{j})
\end{align}

```python
g = 9.8

F1 = lambda t, v: -g*pow(t,0) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
v0 = 0 # Initial Condition

# Explicit Euler Method
v = np.zeros(len(t))
v[0] = v0

for i in range(0, len(t) - 1):
    v[i + 1] = v[i] + h*F1(t[i], v[i])

plt.figure(figsize = (12, 8))
plt.plot(t, v, 'bo--', label='Aproximada')
plt.plot(t, -g*t, 'g', label='Exacta')
plt.title('Método de Euler y caída libre (velocidad)')
plt.xlabel('t (s)')
plt.ylabel('v(t) (m/s)')
plt.grid()
plt.legend(loc='lower left')
plt.show()
```

```python
g = 9.8

F2 = lambda t, x: -g*t # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
x0 = 10 # Initial Condition

# Explicit Euler Method
x = np.zeros(len(t))
x[0] = x0

for i in range(0, len(t) - 1):
    x[i + 1] = x[i] + h*F2(t[i], x[i])

plt.figure(figsize = (12, 8))
plt.plot(t, x, 'ro--', label='Aproximada')
plt.plot(t, 10-0.5*g*pow(t,2), color = 'black', label='Exacta')
plt.title('Método de Euler y caída libre (posición)')
plt.xlabel('t (s)')
plt.ylabel('y(t) (m)')
plt.grid()
plt.legend(loc='lower left')
plt.show()
```

```python
x0 = 10 # Initial Condition

# Trapezoidal
xt = np.zeros(len(t))
xt[0] = x0

for i in range(0, len(t) - 1):
    xt[i + 1] = xt[i] + (h/2)*(F2(t[i], xt[i]) + F2(t[i+1], xt[i+1]))

plt.figure(figsize = (12, 8))
plt.plot(t, xt, 'co--', label='Aproximada')
plt.plot(t, 10-0.5*g*pow(t,2), color = 'black', label='Exacta')
plt.title('Método trapezoidal y caída libre (posición)')
plt.xlabel('t (s)')
plt.ylabel('y(t) (m)')
plt.grid()
plt.legend(loc='lower left')
plt.show()
```

Ejemplo 3:

Péndulo simple + ángulos pequeños.

\begin{equation}
\frac{dS}{dt}=
\begin{pmatrix}
0 & 1 \\
-\frac{g}{L} & 0
\end{pmatrix}
S(t),\,
S(t) = \begin{pmatrix}
\theta(t) \\
\theta^{(1)}(t)
\end{pmatrix}
\end{equation}

Matriz asociada: $\mathcal{M} = \begin{pmatrix}
0 & 1 \\
-\frac{g}{L} & 0
\end{pmatrix}$.

Como es sabido, $\omega^{2} = g/L$. $\mathcal{M} = \begin{pmatrix}
0 & 1 \\
-\omega^{2} & 0
\end{pmatrix}$.

Método explícito:

\begin{equation}
S_{j+1} = 
\begin{pmatrix}
1 & h \\
-\omega^{2}h & 1
\end{pmatrix} S_{j}
\end{equation}


Método implícito:

\begin{equation}
S_{j+1} = 
\begin{pmatrix}
1 & -h \\
\omega^{2}h & 1
\end{pmatrix}^{-1} S_{j}
\end{equation}

Método trapezoidal:

\begin{equation}
S_{j+1} = 
\begin{pmatrix}
1 & -h/2 \\
\omega^{2}h/2 & 1
\end{pmatrix}^{-1}
\begin{pmatrix}
1 & h/2 \\
-\omega^{2}h/2 & 1
\end{pmatrix} S_{j}
\end{equation}

Condiciones del problema: $\omega^{2} = 4$. $\theta_{0} = \theta(0) = 1$. $\theta^{(1)}_{0}=\theta^{(1)}(0)=0$.

Solución exacta:

\begin{align}
\frac{d^{2}\theta}{dt^{2}} &= -4\theta \rightarrow\,\theta(t) = Ae^{\lambda t} \\
\frac{d^{2}\theta}{dt^{2}} & = \lambda^{2}e^{\lambda t}\\
\lambda^{2}e^{\lambda t} &= -4e^{\lambda t} \rightarrow\,\lambda^{2} + 4 = 0 \\
\lambda &= \pm 2i \omega \\
\theta(t) &= Ae^{2it} + Be^{-2it}
\end{align}

Condiciones iniciales:

\begin{align}
\theta_{0} &= \theta(0) = 1 \rightarrow 1 = A + B\\
\frac{d\theta}{dt} &= \theta^{(1)}(t) = 2i\left(Ae^{2it} - Be^{-2it}\right) \\
\theta^{(1)}_{0} &= \theta^{(1)}(0) = 0 = 2i(A-B) \rightarrow 0 = A-B
\end{align}

Sistema de ecuaciones:

\begin{align}
A+B &= 1 \\
A-B &= 0 \\
\text{De la segunda ec., }&A=B \\
2A = 1. \text{ Entonces, }A=\frac{1}{2}
\end{align}

Solución completa:

\begin{align}
\theta(t) = Ae^{2it} + Be^{-2it} &= \frac{1}{2}(e^{2it} + e^{-2it}) \\
e^{2it} + e^{-2it} &= 2\cos 2t \\
\theta(t) &= \cos 2t
\end{align}

```python
from numpy.linalg import inv
```

```python
# define step size
h = 0.1
# define numerical grid
t = np.arange(0, 3*np.pi, h)
# oscillation freq. of pendulum
w = 2
s0 = np.array([[1], [0]])

m_e = np.array([[1, h], 
               [-w**2*h, 1]])
m_i = inv(np.array([[1, -h], 
               [w**2*h, 1]]))
m_t = np.dot(inv(np.array([[1, -h/2], 
    [w**2*h/2,1]])), np.array(
      [[1,h/2], [-w**2*h/2, 1]]))

s_e = np.zeros((len(t), 2))
s_i = np.zeros((len(t), 2))
s_t = np.zeros((len(t), 2))

# do integrations
s_e[0, :] = s0.T
s_i[0, :] = s0.T
s_t[0, :] = s0.T

for j in range(0, len(t)-1):
    s_e[j+1, :] = np.dot(m_e,s_e[j, :])
    s_i[j+1, :] = np.dot(m_i,s_i[j, :])
    s_t[j+1, :] = np.dot(m_t,s_t[j, :])
    
plt.figure(figsize = (12, 8))
plt.plot(t,s_e[:,0],'b-')
plt.plot(t,s_i[:,0],'g:')
plt.plot(t,s_t[:,0],'r--')
plt.plot(t, np.cos(w*t), 'k')
plt.ylim([-3, 3])
plt.xlabel('t')
plt.ylabel('$\Theta (t)$')
plt.legend(['Explicit', 'Implicit', \
            'Trapezoidal', 'Exact'])
plt.show()
```

#Método de Runge-Kutta de 4to orden

Segundo orden:

\begin{align}
\frac{dy}{dx} &= F(x,y(x))\\
y_{j+1} &\approx y_{j} + \int_{x_{j}}^{x_{j+1}}F(x,y(x))\,dx \\
&\text{Aproximación en punto medio:} \\
F(x,y(x)) &\approx F(x_{j+1/2},y_{j+1/2}) + (x-x_{j+1/2})\left. \frac{dF}{dx}\right|_{x_{j+1/2}} + \mathcal{O}(h^{2}),\,x_{j+1/2} = x_{j}+h/2,\,y_{j+1/2} = y(x_{j}+h/2)
\end{align}

Al tomar el punto medio, $\int_{x_{j}}^{x_{j+1}}(x-x_{j+1/2})^{2m+1}\,dx = 0$ (función "impar" en intervalo simétrico).

Entonces,

\begin{align}
\int_{x_{j}}^{x_{j+1/2}}F(x_{j},y_{j})\,dx &\approx hF(x_{j+1/2},y_{j+1/2}) + \mathcal{O}(h^{3}) \\
y_{j+1} &= y_{j} + hF(x_{j+1/2},y_{j+1/2}) + \mathcal{O}(h^{3}).
\end{align}

Función $y(x)$ evaluada en $x_{j+1/2}$: 

\begin{equation}
y_{j+1/2} \approx y_{j} + \frac{h}{2}\frac{dy}{dt} = y_{j} + \frac{h}{2}F(x_{j},y_{j}).
\end{equation}

Sea $k_{1} = hF(x_{j},y_{j})$. De este modo,

\begin{align}
y_{j+1} &\approx y_{j} + k_{2} \\
k_{2} &= hF\left(x_{j}+\frac{h}{2},y_{j} + \frac{k_{1}}{2}\right)
\end{align}

Relación para la expansión a 4to orden:

\begin{align}
y_{j+1} &= y_{j} + \frac{1}{6}(k_{1} + 2k_{2} + 2k_{3} + k_{4}), \\
k_{1} &= hF(x_{j},y_{j}), \\
k_{2} &= hF\left(x_{j}+\frac{h}{2},y_{j}+\frac{k_{1}}{2} \right), \\
k_{3} &= hF\left(x_{j}+\frac{h}{2},y_{j}+\frac{k_{2}}{2}\right), \\
k_{4} &= hF(x_{j}+h,y_{j}+k_{3}).
\end{align}


```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline
```

```python
g = 9.8

F2rk = lambda t, xrk: -g*t # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
x0rk = 10 # Initial Condition


xrk = np.zeros(len(t))
xrk[0] = x0rk

for i in range(0, len(t) - 1):
    k1 = h*F2rk(t[i], xrk[i])
    k2 = h*F2rk(t[i] + 0.5*h, xrk[i] + 0.5*k1)
    k3 = h*F2rk(t[i] + 0.5*h, xrk[i] + 0.5*k2)
    k4 = h*F2rk(t[i] + h, xrk[i] + k3)
    xrk[i + 1] = xrk[i] + (1/6)*(k1 + 2*k2 + 2*k3 +k4)

plt.figure(figsize = (12, 8))
plt.plot(t, xrk, 'ro--', label='Aproximada')
plt.plot(t, 10-0.5*g*pow(t,2), color = 'black', label='Exacta')
plt.title('Runge-Kutta 4to orden y caída libre (posición)')
plt.xlabel('t (s)')
plt.ylabel('y(t) (m)')
plt.grid()
plt.legend(loc='lower left')
plt.show()
```

Movimiento armónico simple + Runge-Kutta:

\begin{align}
\ddot{x} &= -\omega^{2}x \\
v &=\frac{dx}{dt}. \text{ Entonces, }\frac{dv}{dt} = \dot{v} = \ddot{x} \\
\dot{v} &= -\omega^{2}x \\
\dot{x} &= v \\
&\text{De igual modo, }\\
\dot{v} &= F_{\text{oa}}(t,x(t),v(t)) \\
F_{\text{oa}}(t,x(t),v(t)) &= -\omega^{2}x
\end{align}

Por Runge-Kutta,

\begin{align}
v_{j+1} &= v_{j} + \frac{1}{6}(k_{1}^{v} + 2k_{2}^{v} + 2k_{3}^{v} + k_{4}^{v}) \\
x_{j+1} &= x_{j} + \frac{1}{6}(k_{1}^{x} + 2k_{2}^{x} + 2k_{3}^{x} + k_{4}^{x})\\
k_{1}^{v} &= h*F_{\text{oa}}(t_{j},v_{j},x_{j}) \\
k_{1}^{x} &= h*v_{j}\\
k_{2}^{v} &= h*F_{\text{oa}}(t_{j}+0.5h, v_{j}+0.5k_{1}^{v},x_{j}+0.5k_{1}^{x}) \\
k_{2}^{x} &= h*(v_{j} + 0.5k_{1}^{v})\\
k_{3}^{v} &= h*F_{\text{oa}}(t_{j}+0.5h, v_{j}+0.5k_{2}^{v},x_{j}+0.5k_{2}^{x}) \\
k_{3}^{x} &= h*(v_{j} + 0.5k_{2}^{v})\\
k_{4}^{v} &= h*F_{\text{oa}}(t_{j} + h, v_{j}+k_{3}^{v}, x_{j}+ k_{3}^{x}) \\
k_{4}^{x} &= h*(v_{j} + k_{3}^{v})
\end{align}

```python
F1_oa = lambda t, v_oa, x_oa: -pow(w,2)*x_oa

w=2
h = 0.1 # Step size
t = np.arange(0, 3*np.pi + h, h) # Numerical grid
x0_oa = 1 
v0_oa = 0
# Initial Conditions

x_oa = np.zeros(len(t))
v_oa = np.zeros(len(t))

x_oa[0] = x0_oa
v_oa[0] = v0_oa

for i in range(0, len(t) - 1):
    k1v = h*F1_oa(t[i], v_oa[i], x_oa[i]) 
    k1x = h*v_oa[i] 
    k2v = h*F1_oa(t[i] + 0.5*h, v_oa[i] + 0.5*k1v, x_oa[i] + 0.5*k1x)
    k2x = h*(v_oa[i] + 0.5*k1v)
    k3v = h*F1_oa(t[i] + 0.5*h, v_oa[i] + 0.5*k2v, x_oa[i] + 0.5*k2x)
    k3x = h*(v_oa[i] + 0.5*k2v)
    k4v = h*F1_oa(t[i] + h, v_oa[i] + k3v, x_oa[i] + k3x)
    k4x = h*(v_oa[i] + k3v)
    v_oa[i + 1] = v_oa[i] + (1/6)*(k1v + 2*k2v + 2*k3v +k4v)
    x_oa[i + 1] = x_oa[i] + (1/6)*(k1x + 2*k2x + 2*k3x +k4x)

plt.figure(figsize = (13, 8))
plt.plot(t, x_oa, 'ro--', label='Aproximada')
plt.plot(t, np.cos(2*t), color = 'black', label='Exacta')
plt.title('Runge-Kutta 4to orden y MAS')
plt.xlabel('$t$ (s)')
plt.ylabel('$x(t)$ (m)')
plt.grid()
plt.legend(loc='upper center')
plt.show()
```

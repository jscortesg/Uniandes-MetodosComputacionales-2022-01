#Derivación numérica

Diferencia finita tipo "forward" (Forward Difference):

Serie de Taylor $@$ de un punto $h$: $y(t+h) = y(t) + h\frac{dy}{dt}+\frac{h^{2}}{2}\frac{d^{2}y}{dt^{2}}+\cdots$

Si h es muy pequeño, $y(t+h)\approx y(t) + h\frac{dy}{dt}$.

Entonces, \begin{equation}\left. \frac{dy}{dt}\right|_{fd}=\frac{y(t+h)-y(t)}{h}.\end{equation}

Error en la aproximación:

Se toma $\left. \frac{dy}{dt}\right|_{fd}=\frac{y(t+h)-y(t)}{h} + \mathcal{O}(h)$.

Trasponiendo términos, $=\frac{y(t+h)-y(t)}{h} = \left. \frac{dy}{dt}\right|_{fd} - \mathcal{O}(h)$.

Error aproximación: \begin{equation}\epsilon_{fd} = - \mathcal{O}(h) \approx -\frac{h}{2}\frac{d^{2}t}{dt^{2}}\end{equation}.


Ejemplo: para un polinomio cuadrático tipo $y(t) = a +bt^{2}$,

$y(t+h) = a + b(t+h)^{2} = a + b(t^{2} + 2bht +h^{2})$

$y(t+h) - y(t) = a + b(t^{2} + 2bht +h^{2}) - a + bt^{2} = 2bht + bh^{2}$.

$\left. \frac{dy}{dt}\right|_{fd} = \frac{2bht + h^{2}}{h} = 2bt + bh \approx 2bt$.

$\epsilon_{fd} = -\frac{h}{2}\frac{d}{dt}\left[\frac{d}{dt}(a+bt^{2})\right] = -\frac{h}{2}\frac{d}{dt}(2bt) = -hb$.

Diferencia central (Central difference):

\begin{equation}
\left. \frac{dy}{dt}\right|_{cd} \equiv \frac{y(t+h/2)-y(t-h/2)}{h}
\end{equation}

Error en la aproximación:

$y(t\pm h/2) = y(t) \pm \frac{h}{2}\frac{dy}{dt} + \frac{h^{2}}{8}\frac{d^{2}y}{dt^{2}}\pm \frac{h^{3}}{48}\frac{d^{3}y}{dt^{3}}\cdots$

\begin{align}y(t+h/2) - y(t-h/2) &= y(t) + \frac{h}{2}\frac{dy}{dt} + \frac{h^{2}}{8}\frac{d^{2}y}{dt^{2}} + \frac{h^{3}}{48}\frac{d^{3}y}{dt^{3}}\cdots \\
&- \left[y(t) - \frac{h}{2}\frac{dy}{dt} + \frac{h^{2}}{8}\frac{d^{2}y}{dt^{2}} - \frac{h^{3}}{48}\frac{d^{3}y}{dt^{3}}\cdots\cdots\right] \\
&=h\frac{dy}{dt}+\frac{h^{3}}{24}\frac{d^{3}y}{dt^{3}}\end{align}

\begin{equation}\epsilon_{cd} = \frac{h^{2}}{24}\frac{d^{3}y}{dt^{3}}\end{equation}

```python
def ford_cos(x, h):

  return (np.cos(x+h)-np.cos(x))/h, -np.sin(x), -0.5*h*np.cos(x)
```

```python
# paso
h = 0.05
# puntos en x
x = np.arange(0, 0.5*np.pi, h) 
# función
y = np.cos(x) 

# forward difference (comando en numpy)
forward_diff = np.diff(y)/h 
# grilla (todo intervalo menos último elemento)
x_diff = x[:-1:] 
# solución exacta
exact_solution = -np.sin(x_diff) 

font = {'size': 20}
plt.rc('font', **font)

# gráfica
plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', \
         label = 'Aproximación FD')
plt.plot(x_diff, exact_solution, \
         label = 'Solución exacta')
plt.legend()
plt.show()

# error máximo entre derivación numérica y derivada exacta
max_error = max(abs(exact_solution - forward_diff))
print(max_error)
```

#Ejercicio:

Movimiento de un proyectil descrito por las coordenadas

\begin{equation}
\vec{r} = (0.28t + 0.036t^{2})\hat{\imath}+0.019t^{3}\hat{\jmath},
\end{equation}

con $[\vec{r}]=m$ y $[t]=s$.

Encuentre las componentes de la velocidad instantánea en $t=0, 1, 1.5, 5$ s.

Haga la gráfica de las componentes de $\vec{v}$ en el intervalo $0\leq t \leq 5\text{ s}$. Compare con la gráfica original.



Posición en x: $x(t) = 0.28t+0.036t^{2}$.

Posición en y: $y(t) = 0.019t^{3}$.

```python
def velx_FD(t,h):

  def posx(t):
    return 0.28*t+0.036*pow(t,2)

  return (posx(t+h)-posx(t))/h
```

```python
velx_FD(0,0.1)
```

```python
def vely_FD(t,h):

  def posy(t):

    return 0.019*pow(t,3)

  return (posy(t+h)-posy(t))/h
```

```python
vely_FD(0,0.1)
```

```python
listat = [0, 1, 1.5, 5]

for tiempo in listat:
  print("tiempo = ", tiempo, "s, velx = ", round(velx_FD(tiempo,0.1),4), "m/s, vely = ", round(vely_FD(tiempo,0.1),4))
```

```python
def velx_CD(t,h):

  def posx(t):
    return 0.28*t+0.036*pow(t,2)

  return (posx(t+h/2)-posx(t-h/2))/h
```

```python
def vely_CD(t,h):

  def posy(t):

    return 0.019*pow(t,3)

  return (posy(t+h/2)-posy(t-h/2))/h
```

```python
for tiempo in listat:
  print("tiempo = ", tiempo, "s, velx = ", round(velx_CD(tiempo,0.1),4), "m/s, vely = ", round(vely_CD(tiempo,0.1),4))
```

Solución exacta: 

$v_{x} = \frac{dx}{dt} = \frac{d}{dt}(0.28t + 0.036t^{2})$

\begin{equation}
v_{x}=0.28+0.072t
\end{equation}

$v_{y} = \frac{dy}{dt} = \frac{d}{dt}(0.019t^{3})$

\begin{equation}
v_{y} = 0.057t^{2}.
\end{equation}

```python
def vx_exact(t):

  return 0.28+0.072*t
```

```python
def vy_exact(t):

  return 0.057*pow(t,2)
```

```python
plt.figure(figsize = (12, 8))

plt.subplot(1, 2, 1)

h = 0.2

t = np.arange(0, 5, h) 

velx = [velx_FD(t,h) for t in t]

velx_ex = [vx_exact(t) for t in t]

plt.title("$v_{x,FD}$ (m/s),h=0.2")

plt.plot(t, velx, '--', c = 'red')
plt.plot(t, velx_ex, c = 'blue')


plt.subplot(1, 2, 2)

h = 0.2

t = np.arange(0, 5, h) 

vely = [vely_FD(t,h) for t in t]

vely_ex = [vy_exact(t) for t in t] 

plt.title("$v_{y,FD}$ (m/s),h=0.2")

plt.plot(t, vely, 'x-', c = 'black')
plt.plot(t, vely_ex, c = 'orange')
```

```python
plt.figure(figsize = (12, 8))

plt.subplot(1, 2, 1)

h = 0.2

t = np.arange(0, 5, h) 

velxcd = [velx_CD(t,h) for t in t]

velx_ex = [vx_exact(t) for t in t]

plt.title("$v_{x,CD}$ (m/s),h=0.2")

plt.plot(t, velxcd, '--', c = 'red')
plt.plot(t, velx_ex, c = 'blue')


plt.subplot(1, 2, 2)

h = 0.2

t = np.arange(0, 5, h) 

velycd = [vely_CD(t,h) for t in t]

vely_ex = [vy_exact(t) for t in t] 

plt.title("$v_{y,CD}$ (m/s),h=0.2")

plt.plot(t, velycd, 'x-', c = 'black')
plt.plot(t, vely_ex, c = 'orange')
```

#Derivadas de segundo orden

Por el método de CD, 

$\left. \frac{dy}{dt}\right|_{cd}\approx \frac{y(t+h/2)-y(t-h/2)}{h}$.

Tomando una segunda derivada CD,

$\left. \frac{d^{2}y}{dt^{2}}\right|_{cd}\approx \frac{y'(t+h/2)-y'(t-h/2)}{h}$

$y'(t+h/2)=\frac{y(t+h/2+h/2)-y(t+h/2-h/2)}{t+h-(t)} = \frac{y(t+h)-y(t)}{h},$

$y'(t-h/2)=\frac{y(t-h/2+h/2)-y(t-h/2-h/2)}{t-(t-h)} = \frac{y(t)-y(t-h)}{h}.$

En resumen,

\begin{equation}
\left. \frac{d^{2}y}{dt^{2}}\right|_{cd}\approx \frac{y(t+h)+y(t-h)-2y(t)}{h^{2}}.
\end{equation}

#Ejercicio 2

Movimiento de un proyectil descrito por las coordenadas

\begin{equation}
\vec{r} = (0.28t + 0.036t^{2})\hat{\imath}+0.019t^{3}\hat{\jmath},
\end{equation}

con $[\vec{r}]=m$ y $[t]=s$.

Encuentre las componentes de la aceleración instantánea en $t=0, 1, 1.5, 5$ s tomando $h=0.2,0.1,0.05$.

Haga la gráfica de las componentes de $\vec{a}$ en el intervalo $0\leq t \leq 5\text{ s}$. Compare con la gráfica original.


```python
def acelx(t,h):

  def posx(t):
    return 0.28*t+0.036*pow(t,2)

  return (posx(t+h)+posx(t-h)-2*posx(t))/pow(h,2)
```

```python
def acely(t,h):

  def posy(t):

    return 0.019*pow(t,3)

  return (posy(t+h)+posy(t-h)-2*posy(t))/pow(h,2)
```

```python
tlist = [0, 1, 1.5, 5]

listacel1 = [(acelx(t,0.2), acely(t,0.2)) for t in tlist]
listacel2 = [(acelx(t,0.1), acely(t,0.1)) for t in tlist]
listacel3 = [(acelx(t,0.05), acely(t,0.05)) for t in tlist]
```

```python
listacel1
```

```python
listacel2
```

```python
listacel3
```

Aceleración: 

$a_{x} = \frac{d^{2}}{dt^{2}}(0.28t + 0.036t^{2}) = \frac{d}{dt}\frac{d}{dt}(0.28t + 0.036t^{2}) = \frac{d}{dt}(0.28 + 0.072t)$

\begin{equation}
a_{x}(t) = 0.072
\end{equation}

$a_{y} = \frac{d^{2}}{dt^{2}}(0.019t^{3}) = \frac{d}{dt}\frac{d}{dt}(0.019t^{3}) = \frac{d}{dt}(0.057t^2)$

\begin{equation}
a_{y}(t) = 0.114t
\end{equation}

```python
intt1 = np.arange(0,5,0.2)
intt2 = np.arange(0,5,0.1)
intt3 = np.arange(0,5,0.05)

ac1x = [acelx(t,0.2) for t in intt1]
ac2x = [acelx(t,0.1) for t in intt2]
ac3x = [acelx(t,0.05) for t in intt3]

ac1y = [acely(t,0.2) for t in intt1]
ac2y = [acely(t,0.1) for t in intt2]
ac3y = [acely(t,0.05) for t in intt3]
```

```python
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(intt1, ac1x, '-b', label='h=0.2')
ax.plot(intt2, ac2x, '--r', label='h=0.1')
ax.plot(intt3, ac3x, '-xg', label='h=0.05')
ax.axhline(0.072, c='black', label='Exacta', linestyle='--', alpha=0.4)
ax.set_ylim([0.0719,0.0722])
fig.text(0.5, 0.04, 't(s)', ha='center', va='center')
ax.set_title('$a_{x}$ (m/s$^2$)')
leg = ax.legend(prop={"size":20})
```

```python
fig, ax = plt.subplots(figsize=(15,15))
ax.plot(intt1, ac1y, '-b', label='h=0.2')
ax.plot(intt2, ac2y, '--r', label='h=0.1')
ax.plot(intt3, ac3y, '-xg', label='h=0.05')
ax.plot(intt3,[0.114*x for x in intt3], c='black', label='exacta', linestyle='--', alpha=0.4)
fig.text(0.5, 0.08, 't(s)', ha='center', va='center')
ax.set_title('$a_{y}$ (m/s$^2$)')
leg = ax.legend(prop={"size":20})
```
